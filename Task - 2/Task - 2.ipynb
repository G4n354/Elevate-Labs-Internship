```python
# Pandas for Data Manupilation
import pandas as pd
```

```python
# Loading dataset
dataset = pd.read_csv('Processed_Titanic.csv')

# Generate summary statistics
stat = dataset.describe()
print(stat)
```

```python
# Matplotlib & Seaborn for Data Visualization
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Creating histograms and boxplots for numeric features
Types = dataset.select_dtypes(include=['float64', 'int64']).columns
```

```python
# Visualizing all the numeric data
for feature in Types:
    plt.figure(figsize=(12, 6))
    
    # Histogram
    plt.subplot(1, 2, 1)
    sns.histplot(dataset[feature], bins=30, kde=True)
    plt.title(f'Histogram of {feature}')
    
    # Boxplot
    plt.subplot(1, 2, 2)
    sns.boxplot(x=dataset[feature])
    plt.title(f'Boxplot of {feature}')
    
    plt.tight_layout()
    plt.show()
```

```python
# Creating a pairplot for a subset of features
sns.pairplot(dataset[Types])
plt.show()
```

```python
# Calculate the correlation matrix
co_mat = dataset[Types].corr()
```

```python
# Create a heatmap for the correlation matrix
plt.figure(figsize=(12, 8))
sns.heatmap(co_mat, annot=True, fmt=".2f", cmap='coolwarm', square=True)
plt.title('Correlation Matrix')
plt.show()
```
